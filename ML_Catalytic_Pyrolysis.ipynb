```python
#Linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
features = df_new.drop(columns=['Gas yield (%)','Oil yield (%)','Type of plastic','Catalyst used'], axis=1)
Y = df_new['Oil yield (%)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,Y,random_state=10,test_size=0.2)
```

```python
LinearReg = LinearRegression()
LinearReg.fit(x_train,y_train)
```

```python
predictions = LinearReg.predict(x_test)
predictions
```

```python
LinearRegression_R2  =r2_score(y_test,predictions)
LinearRegression_R2
```

```python
#Using a pipeline object to predict oil yield, fit the object using the features and calculate R^2
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
pipe=Pipeline(Input)
pipe
pipe.fit(x_train,y_train)
print("R2 score when using training data for prediction",pipe.score(features,Y))
print("R2 score when using test data for prediction",pipe.score(x_test,y_test))
```

```python
#Support vector regressor
from sklearn.svm import SVR
```

```python
steps=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',SVR(kernel = 'rbf'))]
```

```python
pipe=Pipeline(steps)
pipe
pipe.fit(x_train,y_train)
print("R2 score when using training data for prediction",pipe.score(features,Y))
print("R2 score when using test data for prediction",pipe.score(x_test,y_test))
```

```python
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train,y_train)
```

```python
prediction = regressor.predict(x_test)
prediction
```
